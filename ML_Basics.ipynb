Data Cleaning and Preprocessing
->Removing or filling missing values
->Convert Categorical to numeric
->Drop irrelevant Features(e.g Ticket,Cabin)

```python
#Combine for preprocessing
train['TrainFlag']=1
test['TrainFlag']=0
test['Survived']=-1 #temporary fill
combined=pd.concat([train,test])
print(combined.shape)
```

```python
#Filling missing age and fare
combined['Age'].fillna(combined['Age'].median(),inplace=True)
combined['Fare'].fillna(combined['Fare'].median(),inplace=True)
```

```python
#Fill embarked
combined['Embarked'].fillna(combined['Embarked'].mode()[0],inplace=True)
#Drop Cabin and Ticket (too many missing or irrelevant)
combined.drop(['Cabin','Ticket'],axis=1,inplace=True)
#Encode sex
combined['Sex']=combined['Sex'].map({'male':0,'female':1})
#One hot Encode Embarked
combined=pd.get_dummies(combined,columns=['Embarked'],prefix=['Embarked'])
print(combined.head)
```
